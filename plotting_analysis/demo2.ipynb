```python
F = np.linspace(0, 1, 500)

psucc = F**2 + 2 * F * (1 - F)/3 + 5 * ((1 - F)/3)**2
Fout = (F**2 + ((1 - F)/3)**2) / psucc
```

```python
plt.figure(figsize=(10, 6))
plt.plot(F, Fout, label=r'$F_{out}$', linewidth=2)
plt.plot(F, psucc, label=r'$p_{succ}$', linewidth=2)
plt.xlabel(r'$F_{in}$', fontsize=14)
plt.title(r'$F_{out}$ and $p_{succ}$ as a function of $F_{in}$', fontsize=16)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
